```python
bk1= bk.iloc[:,1:]
```

```python
bk1.columns = ['userID','title','bookRating']
```

```python
bk2=bk1.pivot_table(index='userID',columns='title',values='bookRating').reset_index(drop=True)
bk2
```

```python
bk2.index=bk1['userID'].unique()
bk2
```

```python
bk2.fillna(0,inplace=True)
bk2
```

```python
user_sim=1-pairwise_distances(bk2.values,metric='cosine')
user_sim
```

```python
user_sim2=pd.DataFrame(user_sim)
user_sim2
```

```python
user_sim2.index=bk1['userID'].unique()
user_sim2.columns=bk1['userID'].unique()
user_sim2
```
